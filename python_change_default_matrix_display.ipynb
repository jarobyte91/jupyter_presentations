A common problem in presentations is that too many rows or too few columns are displayed when printing a dataset on the screen.

This can be fixed using the command **pd.options.display.max_rows** and **pd.options.display.max_columns**

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')

df.shape
```

```python
df
```

```python
pd.options.display.max_rows = 10

pd.options.display.max_columns = 100
```

```python
df
```
